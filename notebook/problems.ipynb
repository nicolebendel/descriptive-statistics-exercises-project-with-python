### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
# TODO
import numpy as np

np.random.seed(42)
array_1 = np.random.normal(size = 100)
print(array_1)

# CHI
array_2 = np.random.chisquare(3,100)
print(array_2)

import statistics as stats

#MEAN
print(stats.mean(array_1))
print(stats.mean(array_2))


#MEDIAN 
print(stats.median(array_1))
print(stats.median(array_2))
##

#MODA

print(stats.mode(array_1))
print(stats.mode(array_2))


#MEASURE OF DISPERSRION 
#RANGE

rango_array1 = np.max(array_1) - np.min(array_1)
rango_array2 = np.max(array_2) - np.min(array_2)

print(f'Rango Normal: {rango_array1}')
print(f'Rango Chi: {array_2}')



#VARIANCIA Y DESVIACION ESTANDAR

desviacion1 = np.std(array_1)
varianza1 = np.var(array_1)

desviacion2 = np.std(array_2)
varianza2 = np.var(array_2)

print(f"Normal variance: {varianza1} and std: {desviacion1}")
print(f"Chi variance: {varianza2} and std: {desviacion2}")

#SKEWENS 
from scipy.stats import skew

skew_array1 = skew(array_1)
skew_array2 = skew(array_2)

print(f"Normal skewness: {skew_array1}")
print(f"Chi skewness: {skew_array2}")


#KURTOSIS KURTCOBAIN

from scipy.stats import kurtosis

kurt_array_1 = kurtosis(array_1)
kurt_array_2 = kurtosis(array_2)

print(f"Normal kurtosis: {kurt_array_1}")
print(f"Chi kurtosis: {kurt_array_2}")
````
